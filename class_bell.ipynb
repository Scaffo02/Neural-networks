```python
from scipy.io import loadmat
import numpy as np 
```

```python
dataset = loadmat('test_set.mat')
# Escludi le chiavi interne di MATLAB (__header__, ecc.)
chiavi_utili = [k for k in dataset.keys() if not k.startswith('__')]
# Supponendo che ci sia solo una variabile:
dati = dataset[chiavi_utili[0]].squeeze()
print(dati)
```

```python
C_QS_traing = dati[0:350000,0]
C_RS_traing = dati[0:350000,1]
C_RT_traing = dati[0:350000,2]
C_QT_traing = dati[0:350000,3]
label_traing = dati[0:350000,4]
```

```python
print('Dimensioni del dataset:',C_QS_traing.shape)
```

```python
l =[]

for i in range(C_QS_traing.shape[0]):
    if label_traing[i] == 1:
        l.append(1)
    else:
        l.append(0)
```

```python
print('Numero di stati entangled nel campione:',l.count(1))
print('Percentuale di stati entangled nel campione:',l.count(1)/len(l)*100)
```

```python
label =[]

for i in range(C_QS_traing.shape[0]):
    sum = C_QS_traing[i] + C_RS_traing[i] + C_RT_traing[i] - C_QT_traing[i]
    if sum <= 4:
        label.append(0)
    else:
        label.append(1)

```

```python
correct = []
for i in range(C_QS_traing.shape[0]):
    if label[i] == label_traing[i]:
        correct.append(1)
    else:
        correct.append(0)
```

```python
accuracy = np.sum(correct) / len(correct) * 100
print('Accuratezza:', accuracy, '%')
```

```python
TE = 0
TS = 0
FE = 0
FS = 0

for i in range(C_QS_traing.shape[0]):
    if label[i] == label_traing[i] and label[i] == 1:
        TE += 1
    if label[i] == label_traing[i] and label[i] == 0:
        TS += 1 
    if label[i] != label_traing[i] and label[i] == 1:
        FE += 1
    if label[i] != label_traing[i] and label[i] == 0:   
        FS += 1
       
        
```

```python
print('TE:', TE)
print('TS:', TS)        
print('FE:', FE)
print('FS:', FS)
```
